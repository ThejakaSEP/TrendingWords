### 2. Remove Emojis

```python
#To remove emojis and other characters
def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)
```

```python
emojis_removed = []

for sent in link_removed:
#     emojis_removed.append(deEmojify(tweet))
#     emojis_removed.append(r'(\u00a9|\u00ae|[\u2000-\u3300]|\ud83c[\ud000-\udfff]|\ud83d[\ud000-\udfff]|\ud83e[\ud000-\udfff])','',sent)
    emojis_removed.append(re.sub(r"(\u00a9|\u00ae|[\u2000-\u3300]|\ud83c[\ud000-\udfff]|\ud83d[\ud000-\udfff]|\ud83e[\ud000-\udfff])",'',sent))
    
```

### 5. Tokenization

```python
from nltk.tokenize import word_tokenize
```

```python
tweet_tokens = []

for text in lower_tweets:
    tweet_tokens.append(word_tokenize(text))
```

```python
tweet_tokens
```

```python
from nltk.probability import FreqDist
```

```python
#Checking the Frequency of words
def checkFrequency(ls):
    fdist = FreqDist()
    for sent in ls:
        for word in sent:
            fdist[word] +=1
    return fdist
```

```python
checkFrequency(tweet_tokens)
```

```python
checkFrequency(tweet_tokens).most_common(10)
```

### 6. Stop word Removal

```python
from nltk.corpus import stopwords as sw
```

```python
stopwordsEnglish = sw.words('English')
```

```python
without_stopwords = []

for sent in tweet_tokens:
    arr_temp =[]
    for word in sent:
        if word not in stopwordsEnglish:
            arr_temp.append(word)
    without_stopwords.append(arr_temp)
```

```python
without_stopwords
```

```python
#Function to count totalWords 
def countWords(arr):
    c = 0
    for sent in arr:
        c+=len(sent)
            
    return c
```

```python
print('Total Number of Words before Stop Word Removal: ',countWords(tweet_tokens))
print('Total Number of Words After Stop Word Removal: ',countWords(without_stopwords))
```

```python
#Checking the Frequency of words
fdist = FreqDist()
for sent in without_stopwords:
    for word in sent:
        fdist[word] +=1
fdist
```

### 12. Cosine Similarity

```python
from scipy import spatial

result = 1 - float(spatial.distance.cosine(feature_matrix[0], feature_matrix[1]))
```

```python
result
```

```python
'''In this we will be checking the similarity between each sentences or TF-IDF vectors.
Then we will get cosine distance between 0.5 and 0.7 to being a significant similarity and hence save the relevant 2 sentences to a new list name
similar_sentences. I have selected 0.7 and not 1 because 1 will be mostly repeated or retweeted tweets which 
represents the same and will not give the understanding of cosine similarity for this tutorial'''

similar_sentences = []
for i in range(len(feature_matrix)-1):
    for j in range(i+1,len(feature_matrix)):
        result = 1 - float(spatial.distance.cosine(feature_matrix[i], feature_matrix[j]))
        if 0.5 < result<0.7:
            similar_sentences.append((sentenced[i], sentenced[j]))
            
```

```python
similar_sentences
```

### 13. PCA

```python
def get_similarity(received_vector):
    return cosine_similarity(received_vector)
```

```python
cos_sim = get_similarity(feature_matrix)
```

```python
cos_sim.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
cos_sim_pca = pca.fit_transform(cos_sim)
cos_sim_pca
```

```python
pca.explained_variance_ratio_
```

```python
n = cos_sim_pca.shape[0]
labels = []
```

```python

plt.rcParams["figure.figsize"] = (15,10)

for i in range(n):
    x = cos_sim_pca[i,0]
    y = cos_sim_pca[i,1]
    label = "Text-"+str((i+1))
    labels.append(label)
    plt.scatter(x,y)
#     plt.text(x+0.03,y+0.03,label)
```

### 14. Clustering

```python
from sklearn.cluster import KMeans
```

```python
number_of_clusters = 5
km = KMeans(n_clusters=number_of_clusters)
km.fit(cos_sim_pca)

print(km.labels_)
print(km.cluster_centers_)
```

```python
data = cos_sim_pca
centers = km.cluster_centers_
clusters = km.labels_

plt.scatter(data[:,0],data[:,1],c=clusters)
plt.scatter(centers[:,0],centers[:,1],c='red',marker='*',s=200)
for i in range(n):
    x = data[i,0]
    y = data[i,1]
#     plt.text(x+0.03,y+0.03,labels[i])
```
